##  ngsplot using deeptools
download genome annotation (e.g. for hg19 https://hgdownload-test.gi.ucsc.edu/goldenPath/hg19/bigZips/genes/ - hg19.knownGene)

for hg38 - https://hgdownload-test.gi.ucsc.edu/goldenPath/hg38/bigZips/genes/

```python
!wget https://hgdownload-test.gi.ucsc.edu/goldenPath/hg38/bigZips/genes/hg38.knownGene.gtf.gz
```

```python

%%time
!computeMatrix scale-regions \
 -S ENCFF574NFE.bigWig \
 -R hg38.knownGene.gtf.gz \
 -out matrix.tab.gz \
 -a 2000 -b 2000 \
 --regionBodyLength 4000 \
 --skipZeros \
 --missingDataAsZero
```

```python
!plotHeatmap \
 -m matrix.tab.gz \
 -out result.png \
 --colorMap YlGnBu \
 --regionsLabel 'genes' \
 --heatmapHeight 15 \
 --plotTitle 'ngs_plot' &
```
